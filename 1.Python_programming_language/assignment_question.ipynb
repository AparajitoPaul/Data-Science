**Question 4:** Write a Python program to find the largest of three numbers.

```python
num1 = input("Enter your 1st number: ")
num2 = input("Enter your 2nd number: ")
num3 = input("Enter your 3rd number: ")

l1 = [num1, num2, num3]

def is_number(my_list):
    for i in range(len(my_list)):
        try:
            float(my_list[i])
        except ValueError:
            return False
    return True

if is_number(l1):
    mylist = list(map(float, l1))

    print("You have entered the numbers:", *mylist)
    print(f"Greatest among them is {max(mylist)}")

else:
    print("Invalid input")
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
number = input("Give me a positive integer: ")

def factorial(x):
    result = 1

    while x > 0:
        result *= x
        x -= 1

    return result

if number.isdigit():
    number = int(number)

    print(f"Factorial of {number} is: {factorial(number)}")

else:
    print("invalid input")
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python

n = input("Enter n: ")

def n_sum(a):
    result = 0

    if a.isdigit():
        a = int(a)

        for i in range(1, a + 1):
            result += i

        print(f"Sum of first {a} natural numbers is = {result}")

    else:
        print("Invalid input")

n_sum(n)
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python

def leap_year_check(year: int) -> bool:
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    return False

print(leap_year_check(2004))
print(leap_year_check(2009))
```
