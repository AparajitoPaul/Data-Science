### Assignment 9: Matrix Transposition

Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

```python
def matrix_transpose(matrix):
    transposed=[list(row) for row in list(zip(*matrix))]
    return transposed

mat=[[1,2,3],[4,5,6],[10,11,12]]
matrix_transpose(mat)


```

```python
print(list(zip(*mat)))
```

Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

```python
def flat_list(nl):
    flattened=[item for sublist in nl for item in sublist]
    return flattened
flat_list(mat)
```

```python
def list_reversed(list):
    reversed=list[::-1]
    return reversed
l=[x for x in range(1,10)]
print(l)
print(list_reversed(l))
```

### Assignment 14: List Rotation

Write a function that rotates a list by n positions. Print the original and rotated lists.

```python
def rotate_list(lst,k):
    k=k%len(lst)
    rotated=lst[-k:]+lst[:-k]
    return rotated
l2=[2,3,4,5]
n=3
print(l2)
print(rotate_list(l2,n))
```


### Assignment 15: List Intersection

Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

```python
def intersect_lists(l1,l2):
    result=[]
    for i in l1:
        for j in l2:
            if i==j:
                result.append(i)
    return result

lst1=[x for x in range(1,11)]
lst2=[x for x in range(5,15)]
print(lst1)
print(lst2)
print(intersect_lists(lst1,lst2))



```
