#### List comprehension
* Basic syntax:    `[expression for item in iterable]`

* With conditional logic:  `[expression for item in iterabl if condition]`

* Nested List Comprehension:    `[expression for item1 in iterable1 for item2 in iterable2 ]`

```python
# Basic list comprehension
[x**2 for x in range(10)]
```

```python
# With conditional logic
[x for x in range(10) if x%2!=0]
```

```python
# Mathematical Operations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
#Union
union_set=set1.union(set2)
print(union_set)
```

```python
# Modifying dictionary elements
# Dictionaries are mutable so we can add,update and delete elements
print(students)
```

```python
# Dictionary methods
keys=students.keys()
values=students.values()
items=students.items()
print(keys)
print(values)
print(items)
```

```python
# nested dictionaries
acc_info={
    "acc1":{'name':"Paul","balance":30000,"type_of":"savings"},
    "acc2":{'name':"Roy","balance":30000,"type_of":"current"}
}
acc_info['acc1']['name']
```

```python
# iterating over nested list
for x,y in acc_info.items():
    print(f"{x}:")
    for a,b in y.items():
        print(f"{a}:{b}")
    print("\n")
        
```

```python
# Dictionary Comprehension
squares={x:x**2 for x in range(1,6)}
print(squares)
```

```python
# Conditional Dictionary Comprehension
even_square={x: x**2 for x in range(1,11) if x%2==0}
print(even_square)
```
