### Assignment 9: Working with the `datetime` Module

Use the `datetime` module to print the current date, calculate the date 100 days from today, and determine the day of the week for a given date.

````python
import datetime as dt

# Current Date
today_date=dt.date.today()
print(today_date)

# Date of the day 100 days from today
day_100=today_date+dt.timedelta(days=100)
print(day_100)

# Day of the week 100 days from today
day_100.strftime(format="%A")
````

````python
import random

lst=[x for x in random.sample(range(1,50),5)]
print(lst)

l1=[1,2,3,4,5]
random.shuffle(l1)
print(l1)
````
