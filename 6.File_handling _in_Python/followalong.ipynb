# File handling

## File Operation- Read And Write Files

```python

# Read a whole file
with open ('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # .strip() removes the new line character
```

```python
# Writing a file (overwrting)

with open('example.txt','w') as file:
    file.write("Hello World!\n")
    file.write("This is a new line")
#This basically overwrites the entire cxisting content of example.txt
```

```python
with open ('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# Write a file (Without overwriting)
with open('example.txt','a') as file:
    file.write("\nAppend operation taking place")
```

```python
with open ('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# Write multiple lines into a file using a list of lines
lines=['\nfirst line\n','second line\n','third line\n']
with open('example.txt','a')as file:
    file.writelines(lines)
```

```python
with open ('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
### Binary Files
## Created an empty .bin file called example.bin
hex_bytes = b"\x48\x65\x6c\x6c\x6f\x20\x77\x6f\x72\x6c\x64\x21" # Corresponds to ASCII for "Hello world!"
# Writing in the binary file
with open('example.bin','wb')as file:
    file.write(hex_bytes)
```

```python
# Reading the binary file
with open("example.bin",'rb')as file:
    print(file.read())
```

```python
# Read the content from a source text file and write to a destination text file

with open('example.txt','r')as source_file:
    content=source_file.read()
with open("destination.txt",'w')as destination_file:
    destination_file.write(content)
```

```python
# Read a text file and count the number of lines
line_count=0
with open('example.txt','r') as file:
    for line in file:
        line_count=line_count+1
    print(line_count)
```

```python
# Defining a function to- Read a text file and count the number of line, words and characters

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines=file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines)
        char_count=sum(len(line) for line in lines)
        return line_count,word_count,char_count

file_path="example.txt"
no_of_lines,no_of_words,no_of_chars=count_text_file(file_path)

print(f"LINES:{no_of_lines}\nWORDS:{no_of_words}\nCHARS:{no_of_chars}")


```
