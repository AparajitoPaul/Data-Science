## Inheritance
Inheritance is a fundamental concept in Object Oriented Programming
(OOP)that allows a class to inherit attributes and methods from another class. 

````python
# Example(Single inheritance)
# Parent class
class Car:
    def __init__(self,doors,windows,engine_type):
        self.windows=windows
        self.doors=doors
        self.engine_type=engine_type
    def drive(self):
        print(f"The person will drive the {self.engine_type} car")
````

````python
car1=Car(4,6,'petrol')
````

````python
car1.drive()
````

````python
class Tesla(Car):
    def __init__(self,doors,windows,engine_type,is_selfdriving):
        super().__init__(doors,windows,engine_type)
        self.is_selfdriving=is_selfdriving
    def selfdriving(self):
        print(f"Tesla suuports self driving: {self.is_selfdriving}")
````

````python
tesla1=Tesla(4,5,"Electric",True)
````

````python
tesla1.selfdriving() # Calling method from the child class
````

````python
tesla1.drive() # Calling method from the parent class
````

````python
# Example(Multiple inheritance)
# When a class inherits from multiple classes

# Base class 1
class Animal():
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        print("Subclasses must implement this method")


# Base class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner
````

````python
# Derived Class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} says woof"
````

````python
dog1=Dog("buddy","paul")

dog1.speak()
````

````python
dog1.owner
````

## Polymorphism
 Polymorphism is a core concept of OOP which allows object of different classes to be treated as a sobjects of a common superclass. It provides asingle action in different forms. Polymorphism is typicallly achieved by method overriding and interfaces.

### Method Overriding
Method overriding allows a child class to provide a specific implementation of  amethod that is already defined in the parent class.

````python
# Base class
class Animal:
    def speak(self):
        return "The sound of an animal"
    
# Derived class 1
class Dog(Animal):
    def speak(self):
        return "Woof!"
    
# Derived class 2
class Cat(Animal):
    def speak(self):
        return "Meow!"
 
 
````

````python
dog1=Dog()
cat1=Cat()
print(dog1.speak())
print(cat1.speak())
````

````python
# Function that demonstrates Polymorphism
def animal_speak(animal): # animal here is an object of class Animal
    print(animal.speak())
````

````python
animal_speak(Dog())
````

````python
# Example 2; Polymorphism with Function and Methods

# Base/Parent Claass
class Shape:
    def __init__(self,name):
        self.name=name
    def area(self):
        return "The area of the figure"


# Derived class 1
class Rectangle(Shape):
    def __init__(self,height,width):
        super().__init__("Rectangle")
        self.height=height
        self.width=width

    def area(self):
        return self.width*self.height
    

# Derived class 2
class Circle(Shape):
    def __init__(self,radius):
        super().__init__("Circle")
        self.radius=radius

    def area(self):
        return 3.142*self.radius**2
    


# Functiona that shows polymorphism 
def shape_of_area(shape):
    print(f"The are of {shape.name} is {shape.area()}")
````

````python
rectangle1=Rectangle(12,4)
circle1=Circle(2)
````

````python
shape_of_area(circle1)
````

### Abstract Base Class(ABC)
Abstract base classes are used to define common methods for a grou pf related objects. They can enforce that derived classes implement particular methods, promoting consistency across different implementations.

````python
# Example
from abc import ABC, abstractmethod
````

## Encapsulation & Abstraction
Encapsulation and abstractiona are two fundamental concept/principle of OOP that help in designing robust,maintainable and reusable code. Encapsulation involves bundling data and methods that operatr on the data within a single unit, while absrtaction helps in hiding complex implementation details and exposing only the necessary features.

### Encapsulation
Encapsulation is a concept of wrapping data(variables) and methods(functions) together as a single unit. It restrictsdirect acccess to some of the object's components, which is a means of preventing accidental inference and misuse of the data.

But before jumping into encapsulation let's go though the concepts of public,protected and private variables. These are very important to grasp the encapsulation concept very easily.


````python
## Public, protected and private variable
class Person:
    def __init__(self,name,age):
        self.name=name # Public Variable
        self.age=age # Public Varible

person1=Person("Paul",34)

# Defining a function
def get_name(person):
    return person.name

get_name(person1)
````

````python
dir(person1)  # Notice that public varibale is visible
````

````python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name # Private Variable
        self.__age=age # Private Varible
        self.gender=gender # Public  Variable


        
person1=Person("Paul",34,'Male')
````

````python
dir(person1) # Notice that private variable is not visible but public variable is seen
````

Note that both the private variables are wrangles into `'_Person_age' and `'_Person_name'`

So, a private variable cannot be accessed outside the class

````python
class Person:
    def __init__(self,name,age,gender):
        self._name=name # Protected Variable
        self._age=age # Protected Varible
        self.gender=gender # Public  Variable
````

Only a derived class inheriting a parent class containing protected variable will be able to access the protected varibles.

````python
class Employee(Person): # Derived class
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)
````

#### Encapsulation with Getter and Setter methods

````python
class Person:
    def __init__(self,name,age):
        self.__name=name # Private access modifier/variable
        self.__age=age # Private access modifier/variable
     
    # Getter Method for name
    def get_name(self):
        return self.__name

    # Setter Method for name
    def  set_name(self,name):
        self.__name=name


    # Getter Method for age
    def get_age(self):
        return self.__age
    
     # Setter Method for name
    def  set_age(self,age):
        if age > 0:
            self.__age=age
        else:
            print("Age cannot be negative")
````

### Abstraction

Abstarction is the concept of hiding the complex implementation details and showing only the necessary features of  an object. This helps in reducing programming complexity and effor.

````python
from abc import ABC,abstractmethod
````

````python
## Abstract class always inherits the class ABC

# Abstract base class 
class Vehicle(ABC): # Notice that the class inherits ABC  class
    def drive(self):
        print("The vehicle is used for driving") # Normal method
    @abstractmethod
    def start_engine(self):
        pass


# Child class
class Car(Vehicle):
    def start_engine(self):
        print("car engine started")
````

````python
# Demonstartion of process of abstraction

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()


car1=Car()
operate_vehicle(car1)
````

## Magic Methods

Magic methods in Python, also known as dunder methods(doubleunderscore methods),are special methods that start and end with double underscore. These methods enable you to define the behaviour of objects for bilt_in operations, such as arithmatic operations,comparisons and more

Magic methods are predefines methods in Python that you can override to change the behaviour of your objects. Some common methods includes:

* `__init__`: Initializes a new instance of a class
* `__str__`: Returns a string representation of objects
* `__repr__`: Returns an official string representation of an object
* `__len__`: Returns the length of an object
* `__getitem__`: Gets an item from a container
* `__setitem__`: Sets an item in a container

````python
## Basics Methods

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"


person = Person("Paul", 32)

print(person)
print(repr(person))
````

## Operator Overloading in Python

Operator overloading allows ypou to define the behaviour of opertors(+,-,* etc.) for custom objects. We achieve this by overriding specific magic methods in classes.

#### Common Operator Overloading Magic Methods

| Magic Method | Description |
|-------------|-------------|
| `__add__(self, other)` | Adds two objects using the `+` operator. |
| `__sub__(self, other)` | Subtracts two objects using the `-` operator. |
| `__mul__(self, other)` | Multiplies two objects using the `*` operator. |
| `__truediv__(self, other)` | Divides two objects using the `/` operator. |
| `__eq__(self, other)` | Checks if two objects are equal using the `==` operator. |
| `__lt__(self, other)` | Checks if one object is less than another using the `<` operator. |

...

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, other):
        return Vector(self.x * other, self.y * other)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"


# Example Usage
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = Vector(2, 3)

print(v1 + v2)
print(v1 - v2) 
print(v1 * 3)     

print(v1 == v3) 
print(v1 == v2)    

print(repr(v1))     
````

## Custom Exception( Raise and Throw an excepptin)

````python
# Parent Custom Exception
class Error(Exception):
    pass

# Child Custom Exception
class dobException(Error):
    pass


year = int(input("Enter the DOB year: "))
age = 2024 - year

try:
    if 20 <= age <= 30:
        print("The age is valid, so you can apply for the exam")
    else:
        raise dobException(
            f"Invalid age: {age}. Age must be between 20 and 30."
        )

except dobException as e:
    print("Error:", e)
````

- `Exception` is Python's built-in base class for all exceptions.

```python
class Error(Exception):
    pass
```

- `Error` is a custom exception class that inherits from `Exception`.
- `pass` means the class has no additional attributes or methods.

```python
class dobException(Error):
    pass
```

- `dobException` is another custom exception class.
- It inherits from `Error`.
- It is created specifically to handle age validation errors.
