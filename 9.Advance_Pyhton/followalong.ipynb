## Generators

Generators are a simpler way to create iterators. They use the `yield` keeyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

```python
# Let's define a function square
def square(n):
    for i in range(n):
        return i**2
```

```python
square(3)
```

```python
# But instead if we write 
def square(n):
    for i in range(n):
        yield i**2
```

```python
a=square(3)
```

```python
for i in a:
    print(i)
```

```python
# Also we can use the `next` keyword to object `a`
next(a) #1st execution,1st element,2nd execution 2nd element and so on
```

```python
# Example 2
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen=my_generator()
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

### Practical Example: Reading Large Files

Generators are particularly useful for reading large files because they allow you to process one line at a time without loading the entire file into memory.

```python
# Practical Example : Reading Large Files

def read_large_file(file_path):
    with open(file_path,'r')as file:
        for line in file:
            yield line
```

```python
file_path='large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```

## Decorators

Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

```python
# Suppose we have defined a function

def welcome():
    return "welcome to the advance python section"

# Now if I create a copy of this function
wel=welcome
print(wel) # Notice wel is also a function
```

```python
# Interestingly we can execute the function copy as the original function too
wel()
```

```python
# Even after deletion of the original function!
del welcome
wel()
```

```python
# Closures : A function within function

# Example
def main_welcome():
    msg='Welcome'
    def sub_welcome_method():
        print(msg)
        print("Welcome to the advance python section")
        print("Please learn these concept carefully")
    return sub_welcome_method
```

```python
greet=main_welcome()
```

```python
greet()
```

```python
# Here's another closure example

def main_welcome(func):
    def sub_welcome():
        print("Ok this is pretty interesting!")
        func()
        print("I need to learn this dilligently")
    return sub_welcome()
```

```python
# Function 2
def section_introduction():
    print("In this section we will be going through concepts of iterators, genarators and decorators.")
```

```python
# Without decorator we can just feed Function 2 to manin_welcome
main_welcome(section_introduction)
```

```python
# With decorators however
@main_welcome 
def section_introduction(): # This means that section_introduction fuction is going to get called inside main_welcome function directly
    print("In this section we will be going through concepts of iterators, genarators and decorators.")
```

```python
## Example 2
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called")
        func()
        print("Something is happening after the function  is called")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hellooooooo")
```

```python
say_hello()
```

```python
# Decorators with arguments

def repeat(n):
    def decorator(func):
        def  wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello!")
```

```python
say_hello()
```
